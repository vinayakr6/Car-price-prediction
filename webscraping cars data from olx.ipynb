```python
url = url_df['URL_OLX'].to_list()
```

```python
model = []
Brand = []
variant=[]
fuel=[]
driven_kilometers=[]
transmission = []
number_of_owners=[]
location = []
date_published = []
price = []
```

```python
driver1 = webdriver.Chrome(r'G:\AI Professional\Data Trained\Internship Fliprobo Technologies\Webscraping (Selenium)\chromedriver.exe')
driver1.maximize_window()
```

```python
for a in range(len(url)):
    url_each_car = url[a]
    driver1.get(url_each_car)
    time.sleep(6)
    
    # Manufacturing Year of the car
    name = driver1.find_element(By.XPATH, '//div[@class="_2iMMO"]')
    a = name.text.split()
    model.append(a[-1])

    # Brand Name of the car
    a = name.text.split()
    Brand.append(' '.join(a[:-1]))
    
    # Varinant of the car
    varian = driver1.find_element(By.XPATH, '//div[@class="BxCeR"]')
    variant.append(varian.text)

    # fkt = fuel kilometer transmission
    fkt = driver1.find_elements(By.XPATH, '//div[@class="_3rMkw"]')
    fkt1 = []
    for i in fkt:
        fkt1.append(i.text)
    
    # Fuel type of the car
    fuel.append(fkt1[0])
    
    # Driven kilometers type of the car
    driven_kilometers.append(fkt1[1].replace(' KM',''))
    
    # Transmission type of the car
    transmission.append(fkt1[2])

    # Owner, Location and Date published
    overview = driver1.find_elements(By.XPATH, '//div[@class="_3VRXh"]')
    ov = []
    for j in overview:
        ov.append(j.text)
    
    # Owner of the car
    number_of_owners.append(ov[0][0])
    
    # Location of the car
    location.append(ov[1].split()[-1])
    
    # date published of the car
    date_published.append(ov[-1])
    
    # Price of the car (Target variable)
    pr = driver1.find_element(By.XPATH, '//div[@class="_1uqlc"]')
    price.append(pr.text.replace('₹ ',''))
```
